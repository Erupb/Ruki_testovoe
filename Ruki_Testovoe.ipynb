```python
import requests
import re

def find_phone_numbers(url):
    response = requests.get(url)
    html_content = response.text
    
    phone_regex = re.compile(r'''
        (?:\+7|8)
        \s?[-(]?\s?
        9
        \d{2}
        \s?[-)]?\s?
        \d{3}
        [-]?\s?
        \d{2}
        [-]?\s?
        \d{2}
        |
        (?:\+7|8)
        \s?[-(]?\s?
        9
        \d{2}
        \s?[-)]?\s?
        \d{3}
        [-]?\s?
        \d{2}
        [-]?\s?
        \d{2}
        |
        (?:\+7|8)
        \s?[-(]?\s?
        49\d{1}
        \s?[-)]?\s?
        \d{3}
        [-]?\s?
        \d{2}
        [-]?\s?
        \d{2}
        |
        (?:\+7|8)
        \s?[-(]?\s?
        800
        \s?[-)]?\s?
        \d{3}
        [-]?\s?
        \d{2}
        [-]?\s?
        \d{2}
        |
        (?:\+7|8)
        \s?[-(]?\s?
        9
        \d{2}
        \s?[-)]?\s?
        \d{7}
    ''', re.VERBOSE)
    phone_numbers = phone_regex.findall(html_content)
    
    return phone_numbers

with open('links.txt', 'r') as file:
    for line in file:
        
        found_numbers = find_phone_numbers(line)
        print("Найденные номера телефонов на странице ", line, ":")
        set_res = set(found_numbers) 
        list_res = (list(set_res))

        for item in list_res: 
            print(item) 


```
